```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import folium
import geopandas as gpd

from google.colab import drive
drive.mount('/content/drive')
```

```python
def extract_social_negative_sentences(root_folder):
    # Initialize an empty list to store the data
    data = []

    # Walk through each folder in the root directory
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            print(f"Processing folder: {folder_name}")
            # Walk through each JSON file in the folder
            for file_name in os.listdir(folder_path):
                if file_name.endswith('.json'):
                    file_path = os.path.join(folder_path, file_name)
                    with open(file_path, 'r') as file:
                        json_data = json.load(file)
                        # Extract social negative sentences
                        for sentence, values in json_data.items():
                            if values[0] == 'Social' and values[1].lower() == 'negative':
                                data.append([sentence, folder_name])

    # Create a DataFrame from the collected data
    df = pd.DataFrame(data, columns=['Sentence', 'Folder'])
    return df
```

```python
root_folder = '/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/Areas'

# Extract the data and store it in a DataFrame
df_social_negative = extract_social_negative_sentences(root_folder)
```
