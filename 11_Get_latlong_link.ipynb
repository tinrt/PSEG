```python
import geopy
from geopy.geocoders import Nominatim
import time
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/NJ.csv")
df.head(1)
```

```python
def get_lat_long(municipality):
  geolocator = Nominatim(user_agent="GetLoc")
  location = geolocator.geocode(municipality + ", NJ")

  if location is not None:
    latitude = location.latitude
    longitude = location.longitude
  else:
    latitude = None
    longitude = None

  return latitude, longitude

```

```python
df["Latitude"] = None
df["Longitude"] = None

for i in range(len(df)):
  municipality = df.loc[i, "Municipality"]
  latitude, longitude = get_lat_long(municipality)

  df.loc[i, "Latitude"] = latitude
  df.loc[i, "Longitude"] = longitude

  print(f"Processed {i+1} of {len(df)} municipalities")

  time.sleep(1)
```

```python
df["Bing Search Link"] = "https://www.bing.com/search?q=" + df["Municipality"] + "+New Jersey"
df.head(1)
```

```python
df = df.dropna()
```

```python
df.to_csv("/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/NJ_with_lat_long.csv", index=False)
```
