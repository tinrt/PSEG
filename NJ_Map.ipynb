```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import folium
import geopandas as gpd

from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/label_percentages.csv')
df.head()
```

```python

for column in df.columns[:-1]:
  df[column] = df[column].astype(int)
df.head()
```

```python

columns_to_drop = [col for col in df.columns if 'neutral' in col.lower() or 'none' in col.lower()]

df = df.drop(columns=columns_to_drop)
df.head()
```

```python
df_map=pd.read_csv("/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/NJ_with_lat_long.csv")
df_map.head()
```

```python

df_x = pd.merge(df, df_map, left_on='Folder', right_on='Municipality', how='inner')
df_x
```

```python
geojson_url = 'https://raw.githubusercontent.com/python-visualization/folium/main/examples/data/us-states.json'
gdf = gpd.read_file(geojson_url)
```

```python
import folium

def get_color(row):
    max_value = max(
        row['Social_positive'],
        row['Social_negative'],
        row['Environmental_positive'],
        row['Environmental_negative'],
        row['Governance_positive'],
        row['Governance_negative']
    )
    if max_value == row['Social_positive']:
        return 'green'
    elif max_value == row['Social_negative']:
        return 'red'
    elif max_value == row['Environmental_positive']:
        return 'blue'
    elif max_value == row['Environmental_negative']:
        return 'orange'
    elif max_value == row['Governance_positive']:
        return 'purple'
    elif max_value == row['Governance_negative']:
        return 'darkred'

m = folium.Map(location=[40.0583, -74.4057], zoom_start=8)

folium.GeoJson(gdf).add_to(m)

for idx, row in df_x.iterrows():
    popup_content = (
        f"<strong>{row['Municipality']}</strong><br>"
        f"Social_positive: {row['Social_positive']:.2f}%<br>"
        f"Social_negative: {row['Social_negative']:.2f}%<br>"
        f"Environmental_positive: {row['Environmental_positive']:.2f}%<br>"
        f"Environmental_negative: {row['Environmental_negative']:.2f}%<br>"
        f"Governance_positive: {row['Governance_positive']:.2f}%<br>"
        f"Governance_negative: {row['Governance_negative']:.2f}%"
    )

    popup = folium.Popup(popup_content, max_width=300)

    color = get_color(row)

    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=popup,
        icon=folium.Icon(color=color)
    ).add_to(m)

legend_html = '''
     <div style="position: fixed;
                 bottom: 10px; left: 50%; transform: translateX(-50%); width: 100%;
                 background-color: white; z-index:9999; font-size:14px;
                 border:2px solid grey; padding: 5px; text-align: center;">
     <strong>Legend:</strong>
     <i class="fa fa-map-marker fa-2x" style="color:green"></i> Social_positive
     <i class="fa fa-map-marker fa-2x" style="color:red"></i> Social_negative
     <i class="fa fa-map-marker fa-2x" style="color:blue"></i> Environmental_positive
     <i class="fa fa-map-marker fa-2x" style="color:orange"></i> Environmental_negative
     <i class="fa fa-map-marker fa-2x" style="color:purple"></i> Governance_positive
     <i class="fa fa-map-marker fa-2x" style="color:darkred"></i> Governance_negative
     </div>
     '''

m.get_root().html.add_child(folium.Element(legend_html))

# Display map
m.save('/content/drive/MyDrive/PSEG Research Project/New Jersey Areas/NJ map.html')
```
